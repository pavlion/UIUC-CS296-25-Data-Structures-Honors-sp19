### Puzzle 2: Bar Charts

Create two visualizations, one using `pandas` (`df.[...]`), one using `seaborn`, showing a bar chart displaying the number of faculty with **ONLY** the following academic titles:

- `PROF`
- `ASSOC PROF`
- `ASST PROF`

#### Using a pandas DataFrame:

API: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.bar.html

```python
prof = df[ (df['Academic Title']=='PROF') | (df['Academic Title']=='ASSOC PROF') | (df['Academic Title']=='ASST PROF')]
prof
```

```python
df_prof = pd.value_counts(prof['Academic Title'])
df_prof.plot.bar()
```

#### Using seaborn:

API: https://seaborn.pydata.org/generated/seaborn.barplot.html

*Remember: You can use a DataFrame as the `data` parameter!*
